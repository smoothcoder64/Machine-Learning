```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.impute import SimpleImputer
from impyute.imputation.cs import mice

```

```python
df = pd.read_csv('Train_Data_Final.csv')
```

```python
df['RH'] = 0
for i in range(8000):
    if df['RH_type'][i]=='Dry':
        df['RH'][i] = 0
    elif df['RH_type'][i]=='Ideal':
        df['RH'][i] = 1
    elif df['RH_type'][i]=='Slightly Elevated':
        df['RH'][i] = 2
    elif df['RH_type'][i]=='Elevated':
        df['RH'][i] = 3
    elif df['RH_type'][i]=='High':
        df['RH'][i] = 4
df = df.drop(['RH_type'], axis=1)
```

```python
import seaborn as sns
# Step 1: Data Preprocessing

# Convert 'Date' and 'Time' columns to datetime format
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
sns.lineplot(y='RH', x='DateTime', data=df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.seasonal import seasonal_decompose

# Assuming 'train_data' is the preprocessed and imputed time series data
train_data = df['RH']

# Plot the time series data
plt.figure(figsize=(12, 6))
plt.plot(train_data)
plt.title('Time Series Data')
plt.xlabel('Time')
plt.ylabel('RH Type')
plt.show()

# Plot Autocorrelation Function (ACF)
plt.figure(figsize=(12, 6))
plot_acf(train_data, lags=168)
plt.title('Autocorrelation Function (ACF)')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()

# Seasonal Decomposition
decomposition = seasonal_decompose(train_data, period=24)  # Assuming hourly data with daily seasonality
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot Seasonal Component
plt.figure(figsize=(12, 6))
plt.plot(seasonal)
plt.title('Seasonal Component')
plt.xlabel('Time')
plt.ylabel('Seasonal Amplitude')
plt.show()
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.impute import SimpleImputer



# Step 2: Feature Engineering

# Extract relevant features from datetime
df['DayOfWeek'] = df['DateTime'].dt.dayofweek
df['Month'] = df['DateTime'].dt.month
df['Hour'] = df['DateTime'].dt.hour

# Step 3: Check for outliers and replace with NaN

for column in df.columns:
    if column not in ['Date', 'Time', 'DateTime']:  # Skip non-numeric columns
        z_scores = (df[column] - df[column].mean()) / df[column].std()
        outliers_indices = z_scores[abs(z_scores) > 3].index
        df.loc[outliers_indices, column] = np.nan

# Step 4: Impute missing values using SimpleImputer

imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df.drop(['Date', 'Time', 'DateTime'], axis=1)),
                           columns=df.drop(['Date', 'Time', 'DateTime'], axis=1).columns)


# Now, train ARIMA model on preprocessed and imputed data
# Assuming 'RH_type' is the target variable for ARIMA modeling
target_variable = 'RH'
train_data = df_imputed[target_variable]

# Fit ARIMA model
model = ARIMA(train_data, order=(3,1,0))  # Example order, you may need to tune this
model_fit = model.fit()

# Print summary of ARIMA model
print(model_fit.summary())
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Assuming test_data is the actual values of the target variable for the test set

# Make predictions using the fitted ARIMA model
predictions = model_fit.predict()

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(train_data, predictions)

# Calculate R-squared (R2) score
r2 = r2_score(train_data, predictions)

print("Mean Squared Error (MSE):", mse)
print("R-squared (R2) Score:", r2)
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, r2_score

# Load the preprocessed and imputed data
# Assuming 'df_imputed' contains the preprocessed data with 'DateTime' as the index

# Seasonal Decomposition
decomposition_weekly = seasonal_decompose(df_imputed['RH'], period=168)  # Weekly decomposition
decomposition_monthly = seasonal_decompose(df_imputed['RH'], period=720)  # Monthly decomposition

# Modeling Weekly Trend
trend_weekly = decomposition_weekly.trend.dropna()
model_weekly = SARIMAX(trend_weekly, order=(1, 0, 1), seasonal_order=(1, 1, 1, 168))
results_weekly = model_weekly.fit()

# Modeling Monthly Trend
trend_monthly = decomposition_monthly.trend.dropna()
model_monthly = SARIMAX(trend_monthly, order=(1, 0, 1), seasonal_order=(1, 1, 1, 720))
results_monthly = model_monthly.fit()

# Combine Models
def predict_combined_model(date):
    weekly_prediction = results_weekly.forecast(steps=168).mean()
    monthly_prediction = results_monthly.forecast(steps=720).mean()
    return weekly_prediction + monthly_prediction

# Evaluate and Refine
# Optionally, evaluate the model's performance on a validation set

# Forecasting
# Use the combined model to make predictions for future time periods
# For example, to predict the next 7 days:
next_week_prediction = predict_combined_model(df_imputed.index[-1])

# Print the predictions
print("Next week's prediction:", next_week_prediction)

# Calculate MSE and R2 score
# Get actual values for the last 37 days (7 days for weekly pattern + 30 days for monthly pattern)
actual_values = df_imputed['RH_type'][-888:]

# Predict values for the last 37 days
predicted_values = []
for i in range(888):
    predicted_values.append(predict_combined_model(df_imputed.index[-(i+1)]))

# Reverse the order of predicted values
predicted_values = predicted_values[::-1]

# Calculate MSE and R2 score
mse = mean_squared_error(actual_values, predicted_values)
r2 = r2_score(actual_values, predicted_values)

print("Mean Squared Error (MSE):", mse)
print("R-squared (R2) Score:", r2)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, r2_score

# Load the preprocessed and imputed data
# Assuming 'df_imputed' contains the preprocessed data with 'DateTime' as the index

# Prepare the data
data = df_imputed['RH'].values.reshape(-1, 1)

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
data_normalized = scaler.fit_transform(data)

# Define function to create dataset
def create_dataset(data, time_steps=1):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:(i + time_steps), 0])
        y.append(data[i + time_steps, 0])
    return np.array(X), np.array(y)

# Create the dataset with a time window (e.g., past 30 days)
time_steps = 720  # Change period 30 to 720
X, y = create_dataset(data_normalized, time_steps)

# Reshape input for LSTM [samples, time steps, features]
X = np.reshape(X, (X.shape[0], time_steps, 1))

# Split the data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(units=50))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1)

# Make predictions
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)  # Inverse scaling

# Calculate Mean Squared Error
mse = mean_squared_error(df_imputed['RH'].values[-len(predictions):], predictions)
print("Mean Squared Error (MSE):", mse)

# Calculate R-squared (R2) score
r2 = r2_score(df_imputed['RH'].values[-len(predictions):], predictions)
print("R-squared (R2) Score:", r2)
```
